```python
#ORDER BY: By default pandas will sort things in ascending order. To reverse that, provide ascending=False
#To use frequent query to get DataFrame, I define as follows
def pysqldf(q):
    return sqldf(q,globals())
q="""SELECT *
FROM airport_freq
WHERE airport_ident='KLAX'
ORDER BY type;"""
df=pysqldf(q)
```

```python
df
```

```python
#To get in Descending order:
q="""SELECT *
FROM airport_freq
WHERE airport_ident='KLAX'
ORDER BY type desc;"""
df=pysqldf(q)
```

```python
df
```

```python
#same thing in pandas :
airport_freq[airport_freq.airport_ident=='KLAX'].sort_values('type')
```

```python
#In pandas in descending ordered:
airport_freq[airport_freq.airport_ident=='KLAX'].sort_values('type',ascending=False)
```
