```python
#ORDER BY: By default pandas will sort things in ascending order. To reverse that, provide ascending=False
#To use frequent query to get DataFrame, I define as follows
def pysqldf(q):
    return sqldf(q,globals())
q="""SELECT *
FROM airport_freq
WHERE airport_ident='KLAX'
ORDER BY type;"""
df=pysqldf(q)
```

```python
df
```

```python
#To get in Descending order:
q="""SELECT *
FROM airport_freq
WHERE airport_ident='KLAX'
ORDER BY type desc;"""
df=pysqldf(q)
```

```python
df
```

```python
#same thing in pandas :
airport_freq[airport_freq.airport_ident=='KLAX'].sort_values('type')
```

```python
#In pandas in descending ordered:
airport_freq[airport_freq.airport_ident=='KLAX'].sort_values('type',ascending=False)
```

```python
import pandas as pd
airport=pd.read_csv("airports.csv")
airport_freq=pd.read_csv("airport-frequencies.csv")
runways = pd.read_csv("runways.csv")
```

```python
#IN ... NOT IN in sql should be in WHERE,it gives the values of list in given condition like isin()in pandas
from pandasql import sqldf
def pysqldf(q):
    return sqldf(q,globals())
```

```python
q="""SELECT *
FROM airport
WHERE type IN ('heliport','small_airport');"""
df=pysqldf(q)
```

```python
df
```

```python
#Let's see example of NOT IN
q="""SELECT *
FROM airport
WHERE type NOT IN ('heliport','small_airport');"""
df=pysqldf(q)
```

```python
df
```

```python
# Filtering the condition in Pandas uisng isin
airport[airport.type.isin(['heliport','small_airport'])]
```

```python
#GROUP BY is used how we want to display the data, understand the difference of ORDER BY and GROUP BY. GROUP BY must 
#precede the ORDER BY clause if one is used.
#See below example and try to understand it
q="""SELECT id,ident,type
FROM airport
WHERE type NOT IN ('heliport','small_airport')
GROUP BY type;"""
df=pysqldf(q)
```

```python
df
```

```python
#COUNT function returns the number of rows that matches a specified criteria: See the example and try to understand.
q="""SELECT id,ident,type,COUNT()
FROM airport
WHERE type NOT IN ('heliport','small_airport')
GROUP BY type;"""
df=pysqldf(q)
```

```python
df
```

```python
#AVG(column_name) returns the average value of a numeric column.
q="""SELECT id,ident,type,elevation_ft,AVG(elevation_ft)
FROM airport
WHERE type NOT IN ('heliport','small_airport')
GROUP BY type;"""
df=pysqldf(q)
```

```python
df
```

```python
#SUM() retruns the total sum of the numeric column.
#please see the below example carefully.
q="""SELECT id,ident,type,elevation_ft,SUM(elevation_ft),COUNT(elevation_ft),AVG(elevation_ft)
FROM airport
WHERE type NOT IN ('heliport','small_airport')
GROUP BY type;"""
df=pysqldf(q)
```

```python
df
```

```python
#If WHERE condition is not sufficent we use HAVING COUNT 
# see below example condition should be greater than 200 so first column of above example not printed
q="""SELECT id,ident,type,elevation_ft,SUM(elevation_ft),COUNT(elevation_ft),AVG(elevation_ft)
FROM airport
WHERE type NOT IN ('heliport','small_airport')
GROUP BY type
HAVING COUNT(elevation_ft) > 200;"""
df=pysqldf(q)
```
